```python
import skrf as rf
from skrf.calibration import OnePort
%matplotlib inline
from pylab import *
rf.stylely()

#ideals = rf.read_all('data/ideals/')
#measured = rf.read_all('data/measured/')
#duts = rf.read_all('data/measured/')
ideals = rf.read_all('data/ideals/KirkbySN0092/')
measured = rf.read_all('data/measured/SVA/KirkbySN0092/')
duts = rf.read_all('data/measured/SVA/KirkbySN0092/')

## create a Calibration instance
cal = rf.OnePort(
    ideals = [ideals[k] for k in ['male-short-0092','male-open-0092','male-load-0092']],
    #measured = [measured[k] for k in ['SDRKitsShort','SDRKitsOpen','SDRKitsLoad']],
    measured = [measured[k] for k in ['KirkbyShort','KirkbyOpen','KirkbyLoad']],
    )

caled_duts = [cal.apply_cal(dut) for dut in duts.values()]

ideals['male-short-0092'].plot_s_db()
ideals['male-open-0092'].plot_s_db()
ideals['male-load-0092'].plot_s_db()
#ideals['load'].plot_s_complex()

#duts['SDRKitsShort'].plot_s_db()
#duts['SDRKitsOpen'].plot_s_db()
#duts['SDRKitsLoad'].plot_s_db()

caled_duts[0].plot_s_db()
caled_duts[1].plot_s_db()
caled_duts[2].plot_s_db()

```
